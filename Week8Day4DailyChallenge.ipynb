```python
#first we create a copy of our survey
survey_copy=survey.copy()
```

```python
#We check for missing values
missing_values=survey.isnull().sum()
missing_percentage=(missing_values/len(survey))*100
newDf=pd.DataFrame({'Missing Values':missing_values,'Missing Percentage':missing_percentage})

print("superieur a 50",newDf[newDf["Missing Percentage"]>50])
print("entre 5 et 50 ",newDf[(newDf["Missing Percentage"]<50) & (newDf["Missing Percentage"]>5)])

print("inferieur a 5",newDf[(newDf["Missing Percentage"]<5)& (newDf["Missing Percentage"]>0)])
```

```python
#deals with columns that have more than 50% of missign values drop the columns
survey_copy = survey_copy.drop("other_fuel1", axis='columns')
survey_copy = survey_copy.drop("other_fuel2", axis='columns')
survey_copy = survey_copy.drop("other_fuel3", axis='columns')
survey_copy = survey_copy.drop("wepp_id", axis='columns')
survey_copy = survey_copy.drop("generation_data_source", axis='columns')
survey_copy = survey_copy.drop("owner", axis='columns')
commissioning_year_imputer = SimpleImputer(strategy='median')
survey_copy['commissioning_year'] = commissioning_year_imputer.fit_transform(survey_copy[['commissioning_year']])

year_of_capacity_data_imputer = SimpleImputer(strategy='median')
survey_copy['year_of_capacity_data'] = year_of_capacity_data_imputer.fit_transform(survey_copy[['year_of_capacity_data']])

generation_gwh_2013_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2013'] = generation_gwh_2013_imputer.fit_transform(survey_copy[['generation_gwh_2013']])
generation_gwh_2014_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2014'] = generation_gwh_2014_imputer.fit_transform(survey_copy[['generation_gwh_2014']])
generation_gwh_2015_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2015'] = generation_gwh_2015_imputer.fit_transform(survey_copy[['generation_gwh_2015']])
generation_gwh_2016_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2016'] = generation_gwh_2016_imputer.fit_transform(survey_copy[['generation_gwh_2016']])
generation_gwh_2017_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2017'] = generation_gwh_2017_imputer.fit_transform(survey_copy[['generation_gwh_2017']])
generation_gwh_2018_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2018'] = generation_gwh_2018_imputer.fit_transform(survey_copy[['generation_gwh_2018']])
generation_gwh_2019_imputer = SimpleImputer(strategy='median')
survey_copy['generation_gwh_2019'] = generation_gwh_2019_imputer.fit_transform(survey_copy[['generation_gwh_2019']])
estimated_generation_gwh_2013_imputer = SimpleImputer(strategy='median')
survey_copy['estimated_generation_gwh_2013'] = estimated_generation_gwh_2013_imputer.fit_transform(survey_copy[['estimated_generation_gwh_2013']])
estimated_generation_gwh_2014_imputer = SimpleImputer(strategy='median')
survey_copy['estimated_generation_gwh_2014'] = estimated_generation_gwh_2014_imputer.fit_transform(survey_copy[['estimated_generation_gwh_2014']])
estimated_generation_gwh_2015_imputer = SimpleImputer(strategy='median')
survey_copy['estimated_generation_gwh_2015'] = estimated_generation_gwh_2015_imputer.fit_transform(survey_copy[['estimated_generation_gwh_2015']])
estimated_generation_gwh_2016_imputer = SimpleImputer(strategy='median')
survey_copy['estimated_generation_gwh_2016'] = estimated_generation_gwh_2016_imputer.fit_transform(survey_copy[['estimated_generation_gwh_2016']])
estimated_generation_gwh_2017_imputer = SimpleImputer(strategy='median')
survey_copy['estimated_generation_gwh_2017'] = estimated_generation_gwh_2017_imputer.fit_transform(survey_copy[['estimated_generation_gwh_2017']])
```

```python
#deals with columns that have less than 5% of missign values we drop the rows
survey_copy.dropna(inplace=True)

#We check for missing values
missing_values=survey_copy.isnull().sum()
print(missing_values)
#
```

```python
survey_copy['commissioning_year'] = survey_copy['commissioning_year'].astype(int)
```

```python
#check the differents fuel type
primary_fuel_unique = survey_copy['primary_fuel'].unique()
for fuel in primary_fuel_unique:
    data = survey_copy[survey_copy['primary_fuel'] == fuel]['capacity_mw'].values  # .values â†’ tableau NumPy
    print(f"{fuel}: {np.mean(data)} ")
```

```python
from scipy import stats
primary_fuel_unique = survey_copy['primary_fuel'].unique()
solar=survey_copy[survey_copy['primary_fuel']=="Solar"]['capacity_mw'].values
hydro=survey_copy[survey_copy['primary_fuel']=="Hydro"]['capacity_mw'].values
#hypothese is both have the same average capacity
t_statistic, p_value = stats.ttest_ind(solar, hydro)

# Interpreting the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the  hypothesis")
```
